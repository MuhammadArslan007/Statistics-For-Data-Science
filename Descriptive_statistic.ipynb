### Variance
- tells the degree of spread(how much data point are spread in our full data set) in data set
- calculated by taking average of squared deviation from the mean
- formula sum((xi - mean(x))**2)/N #for population variance
- sum((xi - mean(x))**2)/n-1 # for sample variance
- population is whole group , sample is part of population

```python
data3 = [5,6,6,7,5,8,9,11,2,3]
print(np.mean(data3))
sm = 0
for i in data3:
    d = ((i-np.mean(data3))**2)
    sm = d + sm
print(sm)
sm/len(data3)
```
