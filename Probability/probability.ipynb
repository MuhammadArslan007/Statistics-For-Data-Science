## probability
- probability is the likelihood of the event. (chance of happening)
- is always between 0 and 1
- probability of an event happening = number of ways it can happen / total number of outcomes

    1. mutually exclusive
        - two events cannot occure same time 
        - exp: tossing a coine
    2. non mutually exclusive
        - two events can occure same time
        - exp: withdraw a heart and king card from deck
        
- calculating rule of probability

    1. addivtive rule of probability
        - exp: a single 6-sided die is rolled. what is the probability of rolling a 2 or 3?
        > the events are mutually exclusive. because 2 and 5 cannot appear same time.
        
        > P(A or B) = P(A) + P(B)
        
        > P(2 or 3) = P(2) + P(3)
        
        > = 1/6 + 1/6
        
        > = 2/6
        
        > = 1/3
        
        - experiment: a single card is chosen at random from a standard deck of 52 playing cards. what is the probability of choosing a king or club? (♣ Clubs ♦ Diamonds ♥ Hearts ♠ Spades)
        > the events are non-mutually exclusive. because king and clubs come together.therefore counted twice, so its probability must be subtracted.
        
        > P(A or B) = P(A) + P(B) - P(A and B)
        
        > P(king or club) = P(king) + P(club) - P(king of club)
        
        > = 4/52 + 13/52 - 1/52
        
        > = 16/52
        
        > = 4/13
        
        
        
    2. multiplicative rule of  probability
        - for mutiple time experiment
        
        - exp: toss a coin 3 times
        
        1. independent event
        
        > one event remains unaffected by the occurrence of the other event.
        
        > number of outcome will not reduce
        
        > toss a coin 3 time -> P(H),P(T),P(H) = 1/2, 1/2, 1/2
        
        2. dependent event
        > one event affected by the occurrence of the other event
        
        > when the outcome of the first event affects the outcome of the second so that the probability is changed. number of outcome will keep on reducing.
        
        > find probability to chose 3 aces in 3 time picking a card
        
        > P(3 aces) = 4/52 * 3/51 * 2/50 
        
        > = 24/132,600 
        
        > = 1/5,525
        
        - experiment 1: a coin is tossed and a single 6-sided dice is rolled. find the probability of landing on the head side of the coin and rolling a 3 on the dice.
        > both event are independent because it cannot affect each other.
        
        > if two events A and B are independents then the probability is 
        
        > P(A and B) = P(A) * P(B)
        
        > P(head) = 1/2
        
        > P(3) = 1/6
        
        > P(head and 3) = P(head) * P(3)
        
        > = 1/2 * 1/6
        
        > = 1/12
        
        - experiment 2: a card is chosen at random from a standard deck of 52 playing cards. without replacing it, a second card is chosen. what is the probability that the first card chosen is a queen and the second card chosen is a jack? 
        > both event are dependent
        
        > if two events A and B are independents then the probability is 
        
        > P(A and B) = P(A) * P(B/A)  
        
        > P(B|A) = probability of event B given that event A has already occured  # this is also known as conditional probability , | given that
        
        > P(queen on first pick) = 4/52
        
        > P(jack on 2nd pick given queen on 1st pick) = 4/51
        
        > P(queen and jack) = 4/52 * 4/51
        
        > = 16/2652
        
        > = 4/663

### expected value
> expected value for variable X is the value of X that we would "expect" to get after performing the experiment an infinite number of times

> it is also called the expectation, average or mean value


> mathematically, if random variable X can take x1,x2,x3,...,xn the expected value (EV) is given by:

> EV(X) = x1 * P(X=x1) + x2 * P(X=x2) + x3 * P(X=x3) + ...+ xn * P(X=xn)



- let number of people playing game = 1000
- how many people are expected to  draw 3 red balls, or 2 black balls?
- P(X = 0) = 0.067
- total number of players = 1000
- number of players with 0 red ball = 0.067 * 1000 = 67 people
- similarly
- number of players with 1 red ball = 0.2 * 1000 = 200 people
- number of players with 2 red ball = 0.233 * 1000 = 233 people
- number of players with 3 red ball = 0.333 * 1000 = 333 people
- number of players with 4 red ball = 0.167 * 1000 = 167 people

- how many red balls are we expected on an average in our game?
- total number of red balls = 0 * 67 + 1 * 200 + 2 * 233 + 3 * 333 + 4 * 167 = 2333
- average number of red balls per game = 2333 / 1000 = 2.333
- we can never get 2.333 red balls in one game, as the number of balls will be an integer like 2 or 3

> let X = -10 , 150

> P(X = 150) = P(4 red balls) = 0.167

> P(X = -10) = P(0 red balls) + P(1 red balls) + P(2 red balls) + P(3 red balls) = 0.833

> expected value = (150 * P(X = 150)) + ( -10 * P(X = -10)) 

> = (150 * 0.167) + (-10 * 0.833)

> = 16.72

> on an average, player wins 16.72 rupees

### probability distribution
> X be a random variable that has more than one possible outcome

> if we repeat the experiment many time and plot the probability each possible outcome we get plot that represents the probabilities

> this plot is called the probability distribution(PD)

> two properties

> 1. no negative value 
> 2. all the probability values add up to 1

- > a probability distribution is a form of representation that tells us the probability for all the possible values of X

- there are two type of distribution based on data

1. discrete probability distribution

> these distribution model, the probability of discrete random variable

> discrete random variable: a random variable is discrete it it has finite and countable number of possible outcomes

> exp: tossing a coin. we have only heads or tails in a coin there are no in-between values

> types of discrete probability distribution

> 1. bernoulli distribution
> 2. binomial distribution
> 3. poisson distribution

2. continuous probability distribution

> in these distribution model, probabilities of random variable that have any possible outcome

> - continuous data can take any value within a given range

> - the range may be finite or infinite

> weight of girl can be any value from 54 kgs or 54.1 or 54.2.

> height of students

> types of continuous distribution

> 1. uniform distribution
> 2. normal distribution
> 3. log normal distribution
> 4. exponential distribution

- each probability distribution is defined by a probability function that generalizes the probability of the outcomes.

> - the function is called a probability mass function (PMF) for discrete distriution

> - the function is called a probability density function (PDF) for continuous disrtibution

> - note: the total value of PMF and PDF over the entire domain is always equal to one
